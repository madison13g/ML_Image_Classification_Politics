## Image Processing

```python
import cv2
import urllib.request
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

df = pd.read_csv('./democrats_2023-04-05.csv', index_col=0)
df2 = pd.read_csv('./republicans_2023-04-08.csv', index_col=0)
df = df.append(df2)
print(len(df.index))
```

```python
#df.loc[7, 'pfp']
```

```python
#remove all df rows with no pfp
df = df.dropna(subset=['pfp'])
print(len(df.index))
```

```python
#find finder algorithm
#face_cascade = cv2.CascadeClassifier('./haar/haarcascade_frontalface_alt.xml') #normal
face_cascade = cv2.CascadeClassifier('./haar/haarcascade_eye_tree_eyeglasses.xml') #with glasses

#get pixel function
def url_to_image(url):
    # download the image, convert it to a NumPy array, and then read
    # it into OpenCV format
    resp = urllib.request.urlopen(url)
    image = np.asarray(bytearray(resp.read()), dtype="uint8")
    image = cv2.imdecode(image, cv2.IMREAD_COLOR)
    # return the image
    return image

#detect face function
def facefind(img):
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray, 1.2, 4)
    num = len(faces)
    
    allfaces = []
    for (x, y, w, h) in faces:
        cv2.rectangle(img.astype(np.int32), (x, y), (x+w, y+h), (0, 0, 255), 2)
        allfaces.append(img[y:y + h, x:x + w])
    
    return allfaces, num #array of first face and number of total faces
```

```python
#main loop

numfaces = []
face = []
ids = []

for i, value in enumerate(df['id']):
    try:
        img1 = url_to_image(df.loc[i, 'pfp'])[...,::-1]
        face1 = facefind(img1)
        num = face1[1]
        if num > 0:
            numfaces.append(num)
            face.append(cv2.resize(face1[0][0], (120, 120), interpolation=cv2.INTER_AREA))
            ids.append(value)
    except:
        pass

facedf = pd.DataFrame({'id': ids, 'face': face, 'numfaces': numfaces})

#remerge by ids to get covariates
everything = pd.merge(facedf, df, how='left')

#save it!
import sys
import numpy
numpy.set_printoptions(threshold=sys.maxsize)
everything.to_pickle('croppedfaces.pkl')
```
